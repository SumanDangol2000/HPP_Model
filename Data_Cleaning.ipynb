```python
ds['Title'] = ds['Title'].astype('string')
for index in ds.index:
    #print(index)
    data_title_ = str(ds.loc[index, 'Title'])
    data_title = data_title_.lower().strip().split(' ').copy()
    set_title = set(data_title)
    
    common = set_title.intersection(important_terms_on_property)
    common = list(common)
    if (len (important_terms_on_property.intersection(set_title)) != 0):
        ds.loc[index, 'Title'] = str(common)
       
```

```python
def extract_area_from_string(raw_data):
    raw_data = str(raw_data)
    #print(raw_data)
    if (len(raw_data) > 3):
        try:
            raw_data = raw_data.split(",")
            #print(raw_data[-2].strip().lower())
            return (raw_data[-2].strip().lower())
        except:
            pass
    return ("")
```

```python
def meter_to_feet(data):
    data = str(data)
    try:
        data = data.split(" ")
        if (data[-1] == "meter"):
            return int(int(data[0]) * 3.281)
        else:
            return int(data[0])
    except:
        pass
    return 0
```

```python
first_data = area.str.split('-', expand=True)[0].astype(float)
second_data = area.str.split('-', expand=True)[1].astype(float)
third_data = area.str.split('-', expand=True)[2].astype(float)
fourth_data = area.str.split('-', expand=True)[3].astype(float)

# Perform the calculations
result = (((first_data * 16 + second_data) * 4 + third_data) * 4 + fourth_data) * 21.3906

# Add the result column to the DataFrame
filtered_df['Build Area'] = result

print(filtered_df['Build Area'])
ds['Build Area'] = filtered_df['Build Area']
ds.describe()
```
